```python
# DataFrame Creation
import pandas as pd
data = {
          "Name":["A","B","C"],
          "Marks":[85,90,78]
}
df = pd.DataFrame(data)
print(df)

# df.head() and df.tail()
```

```python
# To determine ths structure of a DataFrame
print(df.shape)
```

```python
# To determine ths structure of a DataFrame
print(df.size)
```

```python
# To display the Coloumn Name
print(df.columns)
```

```python
# To identify the datatype
print(df.dtypes)
```

```python
# To access Individual Column
print(df["Name"])
```

```python
# To access Individual Column
print(df["Marks"])
```

```python
# To access Specific Rows
print(df.loc[0,"Name"])
print(df.loc[0,"Marks"])
print(df.iloc[0,0])
print(df.iloc[2,1])
```

```python
# To add a new Coloumn
df["Grade"] = ["First Class","Distinction","Second Class"]
print(df)
```

```python
# To change the Marks of a Student
df.loc[0,"Marks"] = 88
print(df)
df.iloc[0,1] = 88
print(df)
```

```python
# To delete a Coloumn
# inplace=True ----> Overwrites the Data
df1 = df.drop("Grade",axis=1)
print(df)
print(df1)
```

```python
# To delete a Coloumn
df.drop("Grade",axis=1,inplace=True)
print(df)
```

```python
# Mathematical Operations using Pandas
mean1 = df["Marks"].mean()
max1 = df["Marks"].max()
min = df["Marks"].min()
print(mean1)
print(max1)
print(min)
```

```python
# Filtering of Data using Pandas
print(df[df["Marks"]>80])
```
